## Step 1: Load the Cleaned Data

```python
import pandas as pd
import sqlite3

# Load the cleaned dataset
df = pd.read_csv('../data/processed/clean_sales.csv')

print(f"✅ Data loaded: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nDate range: {df['date'].min()} to {df['date'].max()}")
print(f"\nRegions: {sorted(df['region'].unique())}")
print(f"\nCategories: {sorted(df['product_category'].unique())}")
df.head()
```

## Step 2: Set Up the SQL Engine

This creates an in-memory SQL database from your DataFrame.
Think of it as building a temporary database from your spreadsheet.

Run this cell after loading your data (Step 1 must run first).

```python
# Create an in-memory SQL database from the DataFrame
conn = sqlite3.connect(':memory:')
df.to_sql('sales_transactions', conn, index=False, if_exists='replace')

print("✅ SQL engine ready.")
print("   Table name: sales_transactions")
print(f"   Rows loaded: {df.shape[0]}")
print("\n   You can now run SQL queries against this table.")
```

## Step 3: Define the Query Helper

This creates a helper function `run_query()` that:
1. Takes a SQL string as input
2. Runs it against the in-memory database
3. Returns the results as a formatted DataFrame (table)

Run this cell — you will use `run_query()` throughout the notebook.

```python
def run_query(sql):
    """Run a SQL query and return results as a formatted DataFrame."""
    result = pd.read_sql_query(sql, conn)
    print(f"Query returned {len(result)} rows")
    return result

print("✅ run_query() function is ready.")
print("   Usage: run_query(\"SELECT ... FROM sales_transactions\")")
```

```python
# After running Query 6 — calculate what % of total revenue the top 10 represent
if len(result_6) > 0 and 'total_revenue' in result_6.columns:
    total_all = df['revenue'].sum()
    top_10_revenue = result_6['total_revenue'].sum()
    pct = (top_10_revenue / total_all) * 100
    print(f"Total Q1 revenue (all customers): ${total_all:,.2f}")
    print(f"Top 10 customers revenue:         ${top_10_revenue:,.2f}")
    print(f"Top 10 as % of total revenue:     {pct:.1f}%")
    print()
    if pct > 30:
        print("⚠️  Top 10 represent >30% of revenue — HIGH concentration risk.")
        print("   Losing even 2-3 of these customers would be material.")
    else:
        print("✅ Revenue is reasonably distributed across the customer base.")
```

## Summary: Quick Stats Dashboard

Run this cell for a quick summary of all key metrics — useful for your insight report.

```python
# Quick stats summary — run this to get numbers for your insight report
print("=" * 55)
print("NOVATREND RETAIL — Q1 2024 QUICK STATS")
print("=" * 55)
print(f"Total transactions:          {len(df):,}")
print(f"Total Q1 revenue:            ${df['revenue'].sum():,.2f}")
print(f"Average transaction value:   ${df['revenue'].mean():,.2f}")
print(f"Highest single transaction:  ${df['revenue'].max():,.2f}")
print(f"Unique customers:            {df['customer_id'].nunique()}")
print(f"Unique products:             {df['product_id'].nunique()}")
print(f"Date range:                  {df['date'].min()} to {df['date'].max()}")
print()
print("Revenue by Region:")
regional = df.groupby('region')['revenue'].sum().sort_values(ascending=False)
for region, rev in regional.items():
    pct = (rev / df['revenue'].sum()) * 100
    print(f"  {region:<8}: ${rev:>10,.2f}  ({pct:.1f}%)")
print()
print("Revenue by Category:")
category_rev = df.groupby('product_category')['revenue'].sum().sort_values(ascending=False)
for cat, rev in category_rev.items():
    pct = (rev / df['revenue'].sum()) * 100
    print(f"  {cat:<20}: ${rev:>10,.2f}  ({pct:.1f}%)")
print("=" * 55)
```
